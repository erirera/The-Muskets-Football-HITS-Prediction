```python
wList = [] #Keeping kg as Unit
for val in df_musket.Weight.values:
    val =str(val)
    if 'kg' in val:
        wList.append(float(val[:-2]))
    elif 'lbs'in val:
        wList.append(float(val[:-3])*0.453592) # converting lbs to kg
    else:
          wList.append(float(val))
wList
```

```python
df_musket.Height.unique() #for Height
```

```python
hList = [] #Keeping metre (m) as Unit
for val in df_musket.Height.values:
    val =str(val)
    if 'cm' in val:
        hList.append(float(val[:-2])/100)# converting cm to m 
    elif '\'' in val:
        vals=val.split('\'')
        # ft = vals[0]; and inch = vals[1][:-1]
        hList.append(float(vals[0])*0.3048 + float(vals[1][:-1])*0.0254 ) # converting ft and inches to m
    else:
        hList.append(float(val))
hList
```

```python
dfNew['ht(m)'] = hList
dfNew
```

Task 7 - Create 5 new categorical columns for the Height, Weight, Release Clause, Value and Wage into
which you convert the respective values into clusters/labels as follows
a. Height: Bucket intervals of 0.10 m
b. Weight: Bucket intervals of 10 kg
c. Wage: bucket intervals of 50K
d. Value: bucket intervals of 50M
e. Release Clause: bucket intervals of 50M
Height, Weight, Wage, Value, Release Clause

```python
# Binning for Height (in metres)
Upperbands = [] #taking upperbands
count = 1

while count <= max(dfNew['ht(m)'])/0.10:
    Upperbands.append(0.10*count)
    count += 1
Upperbands
```

```python
dfNew['ht(m)_bins']=pd.cut(x=dfNew['ht(m)'], bins=Upperbands)
dfNew
```

```python
max(dfNew['wt(kg)'])
```

```python
min(dfNew['wt(kg)'])
```

```python
# Binning for Weight (in 10 kg)
Upperbands = [] #taking upperbands
count = 1

while count <= max(dfNew['wt(kg)'])/10:
    Upperbands.append(10*count)
    count += 1
Upperbands
```

```python
dfNew['wt(kg)_bins']=pd.cut(x=dfNew['wt(kg)'], bins=Upperbands)
dfNew
```

```python
# Binning for Wage (in 50K €)
Upperbands = [] #taking upperbands
count = 1

while count <= max(dfNew['Wage(€)'])/50000:
    Upperbands.append(50000*count)
    count += 1
Upperbands
```

```python
dfNew['Wage(€)_bins']=pd.cut(x=dfNew['Wage(€)'], bins=Upperbands)
dfNew
```

```python
# Binning for Value (in 50M €)
Upperbands = [] #taking upperbands
count = 1

while count <= max(dfNew['Value(€)'])/50000000:
    Upperbands.append(50000000*count)
    count += 1
Upperbands
```

```python
dfNew['Value(€)_bins']=pd.cut(x=dfNew['Value(€)'], bins=Upperbands)
dfNew
```

```python
# Binning for Release Clause (in 50M €)
Upperbands = [] #taking upperbands
count = 1

while count <= max(dfNew['Release Clause(€)'])/50000000:
    Upperbands.append(50000000*count)
    count += 1
Upperbands
```

```python
dfNew['Release Clause(€)_bins']=pd.cut(x=dfNew['Release Clause(€)'], bins=Upperbands)
dfNew
```

```python
from sklearn.preprocessing import Normalizer 
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
# convert dataframe to array

dfArr = df_musket.values
X = dfArr[:,:-1]
Y = dfArr[:,-1]

norm = Normalizer().fit(X) # Normalizing the data
normalizedX= norm.transform(X)
test = SelectKBest(score_func=chi2, k=20)
fit = test.fit(normalizedX,Y)
features = fit.transform(normalizedX)
```

```python
fit.scores_
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression ()
res = model.fit(normalizedX,Y)
score = res.score(normalizedX,Y)
score
```
